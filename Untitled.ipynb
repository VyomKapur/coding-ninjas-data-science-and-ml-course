```python
import sqlite3
import numpy as np
import pandas as pd

db = sqlite3.connect('IMDB.sqlite')

df = pd.read_sql_query('select genre.genre, imdb.budget from genre inner join imdb on genre.movie_id = imdb.movie_id', db)

df = df[df.Budget != ""]
df = df[df.genre != ""]
d = {}
# print(df)
for i in df.values:
    d[i[0]] = d.get(i[0], 0) + i[1]

values = np.array(list(d.values()), dtype=float)
keys = np.array(list(d.keys()))
sum = np.sum(values)

for i in range((keys.size)):
    print(keys[i], '%.2f'%(values[i]*100/sum))
```

```python
import sqlite3
import numpy as np
import pandas as pd

db = sqlite3.connect('IMDB.sqlite')
df = pd.read_sql_query('select genre.genre, earning.worldwide, earning.domestic from genre inner join earning on genre.movie_id = earning.movie_id', db)

df = df[df.Worldwide != ""]
df = df[df.Domestic != ""]
df = df[df.genre != ""]
df['net'] = df.Worldwide + df.Domestic
df.drop(columns=['Worldwide', 'Domestic'], inplace=True)

d = {}

# print(df)
for i in df.values:
    d[i[0]] = d.get(i[0], 0) + i[1]

values = np.array(list(d.values()), dtype=float)
keys = np.array(list(d.keys()))
sum = np.sum(values)

for i in range((keys.size)):
    print(keys[i], '%.2f'%(values[i]*100/sum))
```

```python
import requests
import numpy as np
from bs4 import BeautifulSoup

base_url = 'http://quotes.toscrape.com/'
curr_url = 'http://quotes.toscrape.com/page/1/'

response = requests.get(curr_url)

urls = []
while response.status_code == 200:
    data = BeautifulSoup(response.text, 'html.parser')
    authors = data.find_all(class_='author')
    for author in authors:
        if author.text[0] == 'J':
            temp_url = 'http://quotes.toscrape.com/author/'+author.text.replace('.', '-').replace(' ', '-')
            urls.append(temp_url)
    try:
        next_url = data.find(class_='next').a['href']
        curr_url = base_url+next_url
        response = requests.get(curr_url)
    except:
        break
d = {}
```

```python
for url in urls:
    urle = ""
    for i in range(len(urls)):
        if url[i]=='-' and url[i+1] == "-":
            continue
        urle += url[i]
            
    response = requests.get(urle)
    data = BeautifulSoup(response.text, 'html.parser')
    d[data.find("h3").get_text().split('\n')[0]] = data.find(class_="author-born-date").get_text()
```
