```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('0000000000002419_training_ccpp_x_y_train (1).csv')
data.head()
```

```python
x = data.iloc[:, 0:4]
x = np.array(x)
y = data.iloc[:, 4]
y = np.array(y)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled)
x_scaled['5'] = 1
print(x_scaled.head())
x_scaled = np.array(x_scaled)
```

```python
def cost(x, y, beta):
    total_cost = 0
    for i in range(len(x)):
        total_cost += (1/len(x))*((x[i]*beta).sum()-y[i])**2
    return total_cost
```

```python
def gradient_descent(x,y,learn_rate,iterations) :
    num_features = len(x.T)
    beta = np.zeros(num_features)
    for i in range(iterations) :
        slopes = np.zeros(num_features)
        for j in range(len(x)) :
            for k in range(num_features) :
                slopes[k] += -2/len(x) * (y[j] - (beta*x[j]).sum()) * x[j][k]
        beta = beta - learn_rate * slopes
        print(i+1,': Cost =',cost(x,y,beta))
    return beta
```

```python
beta = gradient_descent(x_scaled, y, 0.4, 1300)
```

```python
test = pd.read_csv("0000000000002419_test_ccpp_x_test (1).csv")
test.head()
test = scaler.fit_transform(test)
test = pd.DataFrame(test)
test['5'] = 1
```

```python
test.head()
```

```python
y_pred = test.dot(beta)
```

```python
y_pred = pd.DataFrame(y_pred)
```
